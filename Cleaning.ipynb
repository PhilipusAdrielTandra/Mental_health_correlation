```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("Data Clean.csv", delimiter=";")
# df.info()
# len(df)
```

```python
column_mapping = {'B1R1': 'province',
 'B4K7': 'gender',
 'B4K10THN': 'age',
 'B4K11': 'education',
 'B4K12': 'work_status',
 'B10E01':'injured_past_year',
 'B10E02A':'head_injury',
 'B10E02B':'chest_injury',
 'B10E02C':'back_injury',
 'B10E02D':'stomach_injury',
 'B10E02E':'upper_body_injury',
 'B10E02F':'lower_body_injury',
 'B10E03A':'bruises',
 'B10E03B':'cuts',
 'B10E03C':'sprains',
 'B10E03D':'broken_bones',
 'B10E03E':'severed_limbs',
 'B10E03F':'eye_injury',
 'B10E03G':'brain_damage',
 'B10E03H':'internal_damage',
 'B10E03I':'burns',
 'B10E03J':'other',
 'B10E04A':'impaired',
 'B10E04B':'missing_body_parts',
 'B10E04C':'permanent_injury',
 'B10E05':'place_of_injury',
 'gme':'emotional_mental_health_disorder',
 "Bobot_Normal": "weight_normal"
}

df.rename(columns=column_mapping, inplace=True)
```

```python
df.head()
```

```python
injury_columns = [
    "head_injury", "chest_injury", "back_injury", "stomach_injury",
    "upper_body_injury", "lower_body_injury", "bruises", "cuts", "sprains", "broken_bones",
    "severed_limbs", "eye_injury", "brain_damage", "internal_damage", "burns", "other",
    "impaired", "missing_body_parts", "permanent_injury"
]

df["has_injury"] = df[injury_columns].sum(axis=1) > 0 
df["has_injury"] = df["has_injury"].astype(int)  


group1 = df[df["has_injury"] == 1]["emotional_mental_health_disorder"]
group2 = df[df["has_injury"] == 0]["emotional_mental_health_disorder"]

t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Significant difference: Injuries are associated with mental health disorders.")
else:
    print("No significant difference: Injuries do not significantly affect mental health disorders.")
```
